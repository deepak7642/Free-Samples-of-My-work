In Football matches, points are awarded based on the outcome of each match.

A win being three (3) points, a draw one (1) point and a lose zero (0) point. At the Champions League group stage, each team plays a total of 6 matches intending to eliminate teams with lower points. Only the top two (2) teams get qualified in each group.


For this reasons teams with about **4 wins or more at the group stage are more likely to have qualified**, and teams with **4 losses are more likely to have been disqualified**. It is on this basis that we would elect the teams with the most wins and losses.


The teams with the most wins in the group stages, therefore, are **Liverpool FC, AFC Ajax, and FC Bayern Munchen with 6 wins. Real Madrid CF and Juventus with 5 wins and finally, Manchester City FC and Chelsea FC with 4 wins**.

* *Let's check that if they all qualified.*

```python
teams_with_most_wins = df[df['Wins'].isin(df['Wins'].nlargest(6))]

qualified_teams = df.loc[teams_with_most_wins.index, ['Team Name', 'Qualified']]

print(qualified_teams.to_markdown())
```

* The same logic works for the losing teams. Teams with more than 3 losses are more likely to have been disqualified. This team includes **AC Milan, Besiktas JK, Malmo FF, FC Dynamo Kyiv, FC Shakhtar Donetsk, and Club Brugge.**

```python
teams_with_most_losses = df[df['Loses'].isin(df['Loses'].nlargest(6))]

qualified_teams = df.loc[teams_with_most_losses.index, ['Team Name', 'Qualified']]

print(qualified_teams.to_markdown())
```

***In summary, the rate at which teams win and lose could present information about team qualification based on the competition rules. But this doesn't quite give us insight into the team performance.***

```python
print(df.loc[:, ['Team Name', 'Possession (%)']].sort_values(by=['Possession (%)'], ascending=False).to_markdown())
```

Looking at the overall possession of each team, we see that the possession stats might not be the best factor to determine offensive teams from defensive teams. At least not all on its own.

For those familiar with the sport. You know teams like Manchester City FC (Possession: 57.67%) & FC Bayern Munchen (Possession: 57%) are very offensive teams, but their overall possession isn't reflecting that. This implies that we would have to merge the possession variable to another offensive variable to define offensive teams.

Offensive teams generally have greater passing accuracy because they keep opposing teams in defensive positions. They do this by positioning their attacking force in advanced manners. They also have greater ball recoveries, because once offensive teams lose their possessions, they've enough attacking forces to quickly press and recover the ball as soon as possible.

This makes them (offensive teams) vulnerable defensively to counter attacks though.

```python
fig, ax = plt.subplots(figsize=(8, 5))
df[['Possession (%)', 'Passing accuracy (%)', 'Balls recovered']].plot(kind='density', ax=ax)

# Add grid lines
ax.grid(True)

plt.show()
```

The first thing we notice is the fact that all teams fall in between a tight possession and passing accuracy range. Passing accuracy less skew than possession. We can deduce this from the little skew in both the possession and passing accuracy density from the graph above.

This implies that even though these factors are good, the number of ball recoveries might be the most efficient factor in determining offensive teams. Alongside possession and passing accuracy.

* Let's create a new column that measures offensive and defensive teams. For simplicity, we would convert the ball recovered column to percentage and get the average sum of their percentages to know the most offensive teams in the Champions League.

**Note:** This might not be the most appropriate formula, but it would give us a good enough result.

```python
# Calculate the offensive ratio for each team

df['Offensive Ratio'] = (df['Goals'] / df['Goals conceded'])

# Sort the DataFrame by offensive ratio

df = df.sort_values('Offensive Ratio', ascending=False)

print(df.loc[:, ['Team Name', 'Offensive Ratio']].to_markdown())
```

So the top 5 most offensive teams in the Champions League group stage (according to our formula) are : -  
* FC Bayern München, 
* Real Madrid CF, 
* AFC Ajax,

* Liverpool FC.

The least most offensive teams (or defensive teams) are : -

* Take manchester city and inter milan data separate-out after that we will see who is the favorite to win champions league this time

```python
df_new = df.query("`Team Name` == 'Manchester City FC' or `Team Name` == 'FC Internazionale Milano'")
```

```python
df_new['Best Team'] = np.where(df_new['Team Name'].isin(['Manchester City FC']), 1, 0)
```

```python
print(df_new.to_markdown())
```

```python
X = df_new[['Wins', 'Draws', 'Loses', 'Points', 'Goals', 'Goals conceded', 'Possession (%)', 'Passing accuracy (%)', 'Balls recovered', 'Tackles won', 'Clean sheets', 'Saves', 'Distance covered (km)', 'Assists','Attacks','Clear chances','Penalties scored','Corners taken','Offsides','Runs into attacking third','Runs into key play area','Crossing accuracy (%)','Crosses completed','Crosses attempted','Free-kicks taken','Times in possession','Passes into attacking third','Passes into key play area','Passes into penalty area','Blocks','Attempts conceded off target','Attempts conceded','Saves from penalties','High claims','Low claims','Punches made','Fouls committed','Fouls committed in defensive third','Fouls committed in own half','Fouls suffered']]
y = df_new["Best Team"]

#Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
model = RandomForestClassifier()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

predicted_class_label = predictions[0]

best_team = "Manchester City" if predicted_class_label == 0 else "Inter Milan"

print(f"The model predicted that {best_team} is the favorite team for the Champions League Final which is to be held on June 10th, 2023.")
```
